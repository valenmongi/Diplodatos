**Ejercicio Tablas PESO**

```python
weight = pd.read_csv('wtageinf.csv')
```

```python
weight.head(5)
```

```python
#Tabla peso para Hombres
filtrado_1= weight[weight.Sex==1]
filtrado_1.plot(x='Agemos', y=['P3','P5','P10','P25','P50','P75','P90','P95','P97'])
plt.pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.pyplot.title("Tablas de Peso por percentil - Hombres")
seaborn.despine()

```

```python
#Tabla peso para Mujeres
filtrado_0= weight[weight.Sex==2]
filtrado_0.plot(x='Agemos', y=['P3','P5','P10','P25','P50','P75','P90','P95','P97'])
plt.pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.pyplot.title("Tablas de Peso por percentil - Mujeres")
seaborn.despine()
```

**Ejercicio Paso Fronterizos**

```python
import chardet
import string

string.ascii_letters + string.digits
```

```python
 pasos = pd.read_json('pasos.json', encoding="utf-8")
```

```python
pasos = pd.DataFrame(pasos)
```

```python
pasos_1 = pasos.set_index('name')
```

```python
pasos_2 = pasos_1[pasos_1.index.duplicated(keep=False)]
pasos_2
```

```python
pasos_3 = pasos.drop_duplicates('name')
```

```python
dataset = pd.DataFrame(pasos_3, columns=['name', 'province', 'location'])
```

```python
dataset_24 = dataset.replace(['Cordoba','NEUQUEN'], ['Córdoba','Neuquen'])
```

```python
plt.pyplot.figure(figsize=(25,8))
seaborn.countplot(x='province', data=dataset_24, order = dataset_24['province'].value_counts().index)
plt.pyplot.title("Cantidad de pasos fronterizos por provincia")
seaborn.despine()
```

```python
dataset_1 = pd.DataFrame(dataset_24.groupby('province').size())
dataset_1.rename(columns={0:'Cantidad pasos fronterizos por provincia'}, inplace=True)
```

```python
seaborn.distplot(dataset_1)
```
